```python
import numpy as np
import matplotlib.pyplot as plt

# Data
x = ["python", "c", "c++", "javascript"]
y = [80, 15, 65, 55]
z = [10, 85, 33, 58]

# Set the width of the bars and create x positions
width = 0.4  # Width of each bar
x_positions = np.arange(len(x))  # Numeric positions for each category

# Labels and title
plt.xlabel("Languages", fontsize=12)
plt.ylabel("Percentage of People Learning It", fontsize=12)
plt.title("Fast NUCES")

# Plotting the bars with an offset for each group
plt.bar(x_positions - width / 2, y, width=width, color="pink", edgecolor="black", linewidth=2, label="Popularity (Set XY)")
plt.bar(x_positions + width / 2, z, width=width, color="skyblue", edgecolor="black", linewidth=2, label="Popularity (Set XZ)")

# Add x-tick labels and legend
plt.xticks(x_positions, x)  # Set x-tick labels to original x values
plt.legend()
plt.show()
```

# SCATTER PLOT

```python
dataframe = pd.read_csv("iris.csv")
x = dataframe["sepal.length"]
y = dataframe["petal.length"]
color_map = {'Setosa': '#FF6B6B', 'Versicolor': '#4ECDC4', 'Virginica': '#1A535C'}
colors = dataframe['variety'].map(color_map)
plt.scatter(x, y, c = colors, edgecolor = 'black')
for variety, color in color_map.items():
    plt.scatter([], [], color=color, label=variety) 
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.legend(title='Flower Variety')
plt.show()
```

## Fill_between() function

```python
plt.plot(x,y)
plt.fill_between(x,y)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]
plt.plot(x,y)
plt.fill_between(x=[2,4], y1=2, y2=4)
plt.show()
```

```python
#fill between specifioed range
x=np.arange(1,6)
y=np.arange(1,6)
plt.plot(x, y, color = 'black')
plt.fill_between(x, y, color = '#1A535C', where=(x >= 2) & (x <= 4))
plt.show()
```
